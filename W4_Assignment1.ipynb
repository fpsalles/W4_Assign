### The code below will plot the sine function:

```python
import numpy as np
import matplotlib.pyplot as plt

teta = np.linspace(0, 2*np.pi, 360)
seno = np.sin(teta)

plt.figure()
plt.plot(teta, seno)
plt.grid()
plt.show()
```
